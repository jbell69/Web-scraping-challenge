```python
import requests
from bs4 import BeautifulSoup
from splinter import Browser

import pandas as pd
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#url = input('SENT FROM USER REQUEST')
url3 = 'https://twitter.com/marswxreport?lang=en'
browser.visit(url3)


html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
recent_weather_tweet = soup.findAll("div", class_='js-tweet-text-container')
recent_weather_tweet
```

```python
# Find most recent weather
for a in recent_weather_tweet:
    print(a.text)

# weather = recent_weather_tweet.p.text
# weather
# print(f"Most recent article title is: {Weather}")
```
